```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
epochs = []
accuracies = []
```

```python
epoch = 0
for i in range(10):
    epoch = epoch + 200
    epochs.append(epoch)
    df = pd.read_csv('log_cpu{}.csv'.format(i+1), delimiter=';')
    accuracies.append(df['val_accuracy'][49])
```

```python
plt.xlabel('GAN epochs')
plt.ylabel('Test Accuracy')
plt.plot(epochs, accuracies)
```
